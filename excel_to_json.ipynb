```python
import json
import pandas as pd
```

```python
df = pd.read_excel("data.xlsx")
df.fillna("", inplace = True)
df
```

```python
import ast
json_index = df.to_json(orient ='index') 
json_index = ast.literal_eval(json_index)
json_index = {int(k):v for k,v in json_index.items()}
json_index[0]
with open("data.json", "w") as f:
    json.dump(json_index, f)
```

```python
with open("data.json", "r") as f:
    data = json.load(f)
data['0']
```
